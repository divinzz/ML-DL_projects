```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

data['gender'] = le.fit_transform(data['gender'])

data['part_time_job'] = le.fit_transform(data['part_time_job'])

data['extracurricular_activities'] = le.fit_transform(data['extracurricular_activities'])

career_aspiration_map = {
        'Lawyer': 0, 'Doctor': 1, 'Government Officer': 2, 'Artist': 3, 'Social Network Studies': 4,
        'Software Engineer': 5, 'Teacher': 6, 'Business Owner': 7, 'Scientist': 8,
        'Banker': 9, 'Writer': 10, 'Accountant': 11, 'Designer': 12,
        'Construction Engineer': 13, 'Game Developer': 14, 'Stock Investor': 15,
        'Real Estate Developer': 16
    }

data['career_aspiration'] = data['career_aspiration'].map(career_aspiration_map)

```

```python
x = data[['part_time_job', 'absence_days', 'extracurricular_activities','weekly_self_study_hours', 'math_score','history_score', 'physics_score', 'chemistry_score', 'biology_score','english_score', 'geography_score','total_score','average_score']]
x.head()

y = data['career_aspiration']
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()

xtrain, ytrain = smote.fit_resample(x_train, y_train)
```

```python
print(xtrain.shape)
print(x_test.shape)
print(ytrain.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(xtrain, ytrain)
ypred1=knn.predict(x_test)

from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(xtrain, ytrain)
ypred2=nb.predict(x_test)


from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(xtrain, ytrain)
ypred3=rfc.predict(x_test)


from sklearn.svm import SVC
svc=SVC()
svc.fit(xtrain, ytrain)
ypred4=svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print('Classifier : KNeighborsClassifier')
print('Accuracy : ',accuracy_score (y_test, ypred1),'\n')

print('Classifier :  GaussianNB')
print('Accuracy : ',accuracy_score (y_test, ypred2),'\n')

print('Classifier : RandomForestClassifier')
print('Accuracy : ',accuracy_score (y_test, ypred3),'\n')

print('Classifier : SVC')
print('Accuracy : ',accuracy_score (y_test, ypred4),'\n')
```

```python
models = [knn, nb, rfc, svc]
model_names = ['KNeighbors Classifier', 'GaussianNB', 'Random Forest Classifier', 'Support Vector Machines']


train_scores = []
test_scores = []


for model in models:
    train_scores.append(model.score(xtrain, ytrain))
    test_scores.append(model.score(x_test, y_test))


model_score = pd.DataFrame({
    'Model': model_names,
    'Train Score': train_scores,
    'Test Score': test_scores,
})

model_score
```

```python
import pandas as pd

# Input data
df = {
    'part_time_job': False,
    'absence_days': 2,
    'extracurricular_activities': False,
    'weekly_self_study_hours': 47,
    'math_score': 90,
    'history_score': 86,
    'physics_score': 96,
    'chemistry_score': 100,
    'biology_score': 90,
    'english_score': 88,
    'geography_score': 90
}

# Create DataFrame from input data
df = pd.DataFrame([df])

df['total_score'] = df[[f"{subject}_score" for subject in ['math', 'history', 'physics', 'chemistry', 'biology', 'english', 'geography']]].sum(axis=1)
df['average_score'] = df['total_score'] / len(['math', 'history', 'physics', 'chemistry', 'biology', 'english', 'geography'])

# Make predictions using the existing model
predictions = rfc.predict(df)


career_aspiration_map = {
     0: 'Lawyer', 1: 'Doctor', 2: 'Government Officer', 3: 'Artist', 4: 'Social Network Studies',
    5: 'Software Engineer', 6: 'Teacher', 7: 'Business Owner', 8: 'Scientist',
    9: 'Banker', 10: 'Writer', 11: 'Accountant', 12: 'Designer',
    13: 'Construction Engineer', 14: 'Game Developer', 15: 'Stock Investor',
    16: 'Real Estate Developer'
}

# Map predictions to career aspirations
predicted_career_aspirations = []
for pred in predictions:
    predicted_career_aspirations.append(career_aspiration_map[pred])


total_score = df['total_score'].iloc[0]
average_score = df['average_score'].iloc[0]

print('Total score:', total_score)
print('Average score:', average_score)
print('Predicted Career Aspirations:', predicted_career_aspirations)
```
